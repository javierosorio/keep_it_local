```python
# Ensure necessary packages are installed
# !pip install deep-translator pandas

import pandas as pd
from deep_translator import GoogleTranslator

# Load the data
file_path = '/home/u23/afraaalshammari/Transl/SpaD/New_IR_Data/Round2/TransTask1/Deep/en_annotations_binary_complete.tsv'
data = pd.read_csv(file_path, sep='\t')

# Convert all entries in the 'es' column to strings
data['en'] = data['en'].astype(str)

# Initialize the Google Translator
translator = GoogleTranslator(source='en', target='es')

# Function to split text into chunks of a specified maximum length
def split_text(text, max_length=5000):
    return [text[i:i+max_length] for i in range(0, len(text), max_length)]

# Translate each text in the 'es' column
translated_texts = []
for index, text in enumerate(data['en']):
    try:
        # Debugging: Print the text being translated
        print(f"Translating text {index}: {text}")
        
        # Split text if it exceeds the maximum length
        if len(text) > 5000:
            chunks = split_text(text)
            translated_chunks = [translator.translate(chunk) for chunk in chunks]
            translated_text = ' '.join(translated_chunks)
        else:
            translated_text = translator.translate(text)
        
        # Ensure any remaining Spanish characters are translated
        while any('\u0600' <= c <= '\u06FF' for c in translated_text):
            translated_text = translator.translate(translated_text)
        
        # Debugging: Print the translated text
        print(f"Translated text {index}: {translated_text}")
        
        translated_texts.append(translated_text)
    except Exception as e:
        print(f"Error translating text {index}: {text}. Error: {e}")
        translated_texts.append(text)  # Append the original text if translation fails

# Replace the original text with the translated text
data['en'] = translated_texts

# Save the translated data to a new TSV file
translated_file_path = '/home/u23/afraaalshammari/Transl/SpaD/New_IR_Data/Round2/TransTask1/Deep/Deep_En_ES_Completed_Translated.tsv'
data.to_csv(translated_file_path, sep='\t', index=False)

print("Translation completed and saved to", translated_file_path)
```
